```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft2,fftshift
```

```python
def oversample(array,OS):
    
    dim = array.shape[0]
    size = int(dim*OS)
    
    box = np.zeros([size,size],dtype='complex128')
    
    box[int(box.shape[0]/2)-int(dim/2):int(box.shape[0]/2)+int(dim/2),int(box.shape[0]/2)-int(dim/2):int(box.shape[0]/2)+int(dim/2)] = array
    
    return box

def LoadABCD(wl):
    
    # Thin Lens to Focus
    dist = 10.
    efl = 10.
    
    return np.array([[1-dist/efl,0,dist,0],
                     [0,1-dist/efl,0,dist],
                     [-1/efl,0,1,0],
                     [0,-.99999/efl,0,1]])

def HuygensFresnel(ABCD,field,wl,defocus=0):
    
    
    npix = field.shape[0]
    
    # I thought the u parameter would be the detector - that does not appear to be how it is operating
    # These are called the "position vectors in the input and output planes"
    
    x = np.linspace(-0.5,0.5,npix) # changing this also appears to apply defocus whyy
    r_norm = np.max(x)
    u = np.linspace(-5e-4,5e-4,npix) # changing this appears to apply defocus, which is w r o n g
    u_norm = np.max(u)
    
    x /= r_norm
    u /= u_norm
    
    x1,y1 = np.meshgrid(x,x)
    x2,y2 = np.meshgrid(u,u)
    
    if defocus != 0:
        print('applying {} waves of defocus'.format(defocus))
        defocus *= wl
        field *= np.exp(1j*2*np.pi/wl*defocus*(x1**2 + y1**2))
    
    array_to_display = field
    array_to_display[(x1**2 + y1**2) >= 1 ] = 0
    
    plt.figure(figsize=[10,5])
    plt.subplot(121)
    plt.title('Entrance Pupil Field')
    plt.imshow(np.abs(array_to_display))
    plt.colorbar()
    plt.subplot(122)
    plt.title('Entrance Pupil Phase')
    plt.imshow(np.angle(array_to_display))
    plt.colorbar()
    plt.show()
    
    print('Ray Transfer Matrix')
    print(ABCD)
    
    A = ABCD[0:2,0:2]
    B = ABCD[0:2,2:4]
    C = ABCD[2:4,0:2]
    D = ABCD[2:4,2:4]
    
    Axx = A[0,0]
    Axy = A[0,1]
    Ayx = A[1,0]
    Ayy = A[1,1]
    
    Dxx = D[0,0]
    Dxy = D[0,1]
    Dyx = D[1,0]
    Dyy = D[1,1]
    
    amp = 1j/(np.linalg.det(B)**(1/2) * wl)
    
    Binv = np.linalg.inv(B)
    
    Bxx = Binv[0,0]
    Bxy = Binv[0,1]
    Byx = Binv[1,0]
    Byy = Binv[1,1]
    
    O = (C - A @ Binv @ D)
    Oxx = O[0,0]
    Oxy = O[0,1]
    Oyx = O[1,0]
    Oyy = O[1,1]
    
    # Program Parabolic Phase outside of integrand
    outside_x = ((x2*Dxx + y2*Dyx)*Bxx + (x2*Dxy + y2*Dyy)*Byx)*x2
    outside_y = ((x2*Dxx + y2*Dyx)*Bxy + (x2*Dxy + y2*Dyy)*Byy)*y2
    outside = outside_x + outside_y
    
    # Program Parabolic Phase inside of integrand
    inside_x = ((x1*Bxx + y1*Byx)*Axx + (x1*Bxy + y1*Byy)*Ayx)*x1
    inside_y = ((x1*Bxx + y1*Byx)*Axy + (x1*Bxy + y1*Byy)*Ayy)*y1
    inside = inside_x + inside_y
    
    # Crossed term
    cross = ((x2*Oxx + y2*Oyx)*x1 + (x2*Oxy + y2*Oyy)*y1)
    
    # The Fourier Kernel Terms
    fourier = -2*(Bxy*x1*y2 + Byx*x2*y1)
    
    # Assemble
    phase = np.exp(-1j*np.pi/wl * (inside + fourier + cross  + outside))
    
    field = oversample(field*phase,8)
    
    collins = amp * fftshift(fft2(fftshift(field)))
    
    return collins
```

```python
from numpy.fft import fft2,fftshift
from matplotlib.colors import LogNorm

npix = 512
wl = 1e-6
os = 8

# Create dimensions for 
x = np.linspace(-1,1,npix)
x,y = np.meshgrid(x,x)

array = np.zeros([npix,npix],dtype='complex128')
array[np.sqrt(x**2 + y**2) <= 1] = 1

# Evaluate Field
cut = 128
field = HuygensFresnel(LoadABCD(wl),array,wl,defocus=0)[int(npix*os/2)-cut:int(npix*os/2)+cut,int(npix*os/2)-cut:int(npix*os/2)+cut]
field /= np.max(field)

plt.figure(figsize=[10,5])
plt.subplot(121)
plt.title('Propagated Log10(|Irradiance|)')
plt.imshow(np.abs(field*np.conj(field)),norm=LogNorm())
plt.colorbar()
plt.subplot(122)
plt.title('Propagated Field Phase')
plt.imshow(np.angle(field))
plt.colorbar()
plt.show()
```
